```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2


DATADIR = "C:/Users/PC PRO ULTRA RAPIDE/Desktop/Projects/artificial intelligence/ml/sign-lang/sign-lang-dataset/"

ALPHABETS = ["alif", "ain"]

for alphabet in ALPHABETS:
    path = os.path.join(DATADIR, alphabet)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show()
```

```python
training_data = []

def create_training_data():
    for alphabet in ALPHABETS:
        path = os.path.join(DATADIR, alphabet)
        class_num = ALPHABETS.index(alphabet) # 0 for alif and 1 for ain
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                training_data.append([img_array, class_num])
            except Exception as e:
                pass

create_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
for sample in training_data:
    print(sample[1])
```

```python
X = [] # features set
y = [] # label set

for feature, label in training_data:
    X.append(feature)
    y.append(label)
    
X = np.array(X , dtype=object).reshape(-1, 64, 64, 1)
```

```python
# Saving the data set for later use :)

import pickle

pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
